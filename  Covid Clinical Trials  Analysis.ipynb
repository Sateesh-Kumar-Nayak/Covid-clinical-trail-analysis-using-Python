 #  Handling Missing Values

We will handle missing values with the following strategy:

Drop columns with excessive missing data.
Fill categorical missing values with 'Unknown' or 'Not Applicable'.
Impute numerical columns like Enrollment with the median.

```python
# Dropping columns with more than 90% missing values
columns_to_drop = ['Acronym', 'Results First Posted', 'Study Documents']
covid_clinical_trials_cleaned = covid_clinical_trials.drop(columns=columns_to_drop)
```

```python
# Filling missing values for categorical columns
covid_clinical_trials_cleaned['Phases'].fillna('Not Applicable', inplace=True)
covid_clinical_trials_cleaned['Interventions'].fillna('Unknown', inplace=True)
```

```python
# Filling missing values in the 'Enrollment' column with the median
covid_clinical_trials_cleaned['Enrollment'].fillna(covid_clinical_trials_cleaned['Enrollment'].median(), inplace=True)
```

```python
# Rechecking for missing values
covid_clinical_trials_cleaned.isnull().sum()
```

# Exploratory Data Analysis (EDA)

 # Distribution of Clinical Trial Status

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=covid_clinical_trials_cleaned, y='Status', order=covid_clinical_trials_cleaned['Status'].value_counts().index)
plt.title('Distribution of Clinical Trial Status')
plt.xlabel('Count')
plt.ylabel('Status')
plt.show()
```

The majority of COVID-19 clinical trials are in the "Recruiting" stage, followed closely by "Completed" trials. This suggests that while many trials are still ongoing, a substantial number of trials have already concluded, indicating significant progress in COVID-19 research efforts.

# Distribution of Phases of Clinical Trials

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=covid_clinical_trials_cleaned, y='Phases', order=covid_clinical_trials_cleaned['Phases'].value_counts().index)
plt.title('Distribution of Trial Phases')
plt.xlabel('Count')
plt.ylabel('Phases')
plt.show()
```

Insight: A significant portion of trials are categorized under "Not Applicable," which often includes observational studies or early-phase trials that do not follow traditional clinical trial phases. Among those that are phased, most trials are in Phase 2, indicating that many trials are actively testing treatments for efficacy.

# Distribution of Study Types

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=covid_clinical_trials_cleaned, x='Study Type', order=covid_clinical_trials_cleaned['Study Type'].value_counts().head(2).index)
plt.title('Distribution of Study Types')
plt.xlabel('Study Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Insight: The dataset contains more interventional studies than observational ones. This indicates a larger focus on treatment and intervention testing, with a smaller share of trials focusing on understanding disease behavior without intervention.

# Distribution of Enrollment in Trials

```python
plt.figure(figsize=(10, 6))
sns.histplot(covid_clinical_trials_cleaned['Enrollment'].dropna(), bins=50, kde=True)
plt.title('Distribution of Enrollment in COVID-19 Clinical Trials')
plt.xlabel('Enrollment')
plt.ylabel('Frequency')
plt.xlim(0, 100000)  # Limiting the x-axis for better visualization
plt.show()
```

Insight: Most clinical trials have fewer than 1,000 participants, with a few large outlier trials involving tens of thousands of participants. This suggests that while many trials are conducted on a smaller scale, some major trials have broader participant bases, likely for vaccine testing.

# Top 5 Sponsors

```python
# Extracting the top 5 sponsors
top_sponsors = covid_clinical_trials_cleaned['Sponsor/Collaborators'].value_counts().head(5)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_sponsors.index, y=top_sponsors.values)
plt.title('Top 5 Sponsors in COVID-19 Clinical Trials')
plt.xlabel('Sponsor')
plt.ylabel('Number of Trials')
plt.xticks(rotation=45)
plt.show()
```

Insight: The top sponsors include major pharmaceutical companies, government agencies, and research institutions. Leading sponsors, such as the NIH and AstraZeneca, have been pivotal in driving large numbers of trials, showcasing their leadership in COVID-19 research.

# Gender Eligibility in Trials

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=covid_clinical_trials_cleaned, x='Gender')
plt.title('Distribution of Gender Eligibility in Clinical Trials')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

Insight: Most clinical trials are open to both male and female participants, demonstrating inclusivity in the research process. A small number of trials, however, focus on gender-specific conditions, where eligibility is restricted to either males or females.

# Countries with the Most Trials

```python
covid_clinical_trials_cleaned['Country'] = covid_clinical_trials_cleaned['Locations'].str.split(',').str[-1].str.strip()

top_countries = covid_clinical_trials_cleaned['Country'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Top 10 Countries with the Most Clinical Trials')
plt.xlabel('Number of Trials')
plt.ylabel('Country')
plt.show()
```

# Most Studied Conditions

```python
top_conditions = covid_clinical_trials_cleaned['Conditions'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_conditions.values, y=top_conditions.index)
plt.title('Top 10 Most Studied Conditions in COVID-19 Trials')
plt.xlabel('Number of Trials')
plt.ylabel('Condition')
plt.show()
```

Insight: The United States is the leading country conducting COVID-19 clinical trials, followed by European nations such as France and the UK. This reflects the global nature of COVID-19 research, but also the dominance of wealthier countries in spearheading research efforts.

# Distribution of Outcome Measures

```python
# Here, we will plot the top outcome measures
top_outcomes = covid_clinical_trials_cleaned['Outcome Measures'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_outcomes.values, y=top_outcomes.index)
plt.title('Top 10 Outcome Measures in COVID-19 Clinical Trials')
plt.xlabel('Number of Trials')
plt.ylabel('Outcome Measures')
plt.show()
```

Insight: The primary condition under study is, unsurprisingly, COVID-19 itself. However, related conditions such as pneumonia, respiratory distress, and viral infections are also common, emphasizing the wide array of symptoms and complications linked to the virus.
